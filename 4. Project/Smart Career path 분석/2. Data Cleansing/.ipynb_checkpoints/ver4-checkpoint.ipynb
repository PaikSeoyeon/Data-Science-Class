```python
import numpy as np
import pandas as pd
import string
# import missingno as msno
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import re
import nltk
nltk.download('stopwords')
stopword = nltk.corpus.stopwords.words('english')
wn = nltk.WordNetLemmatizer()
nltk.download('wordnet')
```

```python
df = pd.read_json('dataengineer.json',lines=True)
```

```python
df.columns
```

```python
## nlp에 필요한 function정의
def remove_punct(text):
    text_nopunct = "".join([char for char in text if char not in string.punctuation])
    return text_nopunct
def tokenize(text):
    tokens = re.split('\W+', text)
    return tokens
def remove_stopwords(tokenized_list):
    text = [word for word in tokenized_list if word not in stopword]
    return text
def lemmatizing(tokenized_text):
    text = [wn.lemmatize(word) for word in tokenized_text]
    return text
```

## SKILL

```python
## df 원본은 그대로 두기 위해 prep할 skill만 빼옴
df_sk = df[['Skills & Endorsements','Tools & Technologies','Interpersonal Skills','Other Skills','Industry Knowledge']]
df_sk.head()
```

```python
## skill들을 빼올수 있는 함수 생성
def get_skill(sel):
    skill_list = []
    for i in range(len(sel)):
        skill_list.append(sel[i]['Skills'])
    return skill_list
```

```python
## 바로 apply를 하면 nan은 float type이라서 len가 안먹힌다.
## 각 column을 공통 type으로 맞춰주기 위해 nan값을 찾아 빈리스트로 대체
def nan_to_list(num):
    if num!=num:
        return list()
    else:
        return num
```

```python
## 각 스킬 컬럼에 nan을 빈 list로 대체하여 skill list를 구하기
df_sk['Skill'] = df_sk['Skills & Endorsements'].apply(lambda x:nan_to_list(x)).apply(lambda x:get_skill(x))
df_sk['Tool'] = df_sk['Tools & Technologies'].apply(lambda x:nan_to_list(x)).apply(lambda x:get_skill(x))
df_sk['Personal Skill'] = df_sk['Interpersonal Skills'].apply(lambda x:nan_to_list(x)).apply(lambda x:get_skill(x))
df_sk['Etc Skill'] = df_sk['Other Skills'].apply(lambda x:nan_to_list(x)).apply(lambda x:get_skill(x))
df_sk['Knowledge'] = df_sk['Industry Knowledge'].apply(lambda x:nan_to_list(x)).apply(lambda x:get_skill(x))
```

```python
df_skill = df_sk[['Skill','Tool','Personal Skill','Etc Skill','Knowledge']]
df_skill.head()
```

```python
## 각 skill column에서 모든 skill들 list를 구하는 함수
def get_skill_list(ser):
    ret = []
    for idx in range(len(ser)):
        for i,v in enumerate(ser.iloc[idx]):
            ret.append(v)
    return ret
```

```python
## skill_list는 모든 사람이 가지고 있는 스킬들(중복허용)
skill_list1 = get_skill_list(df_skill['Skill'])
skill_list2 = get_skill_list(df_skill['Tool'])
skill_list3 = get_skill_list(df_skill['Personal Skill'])
skill_list4 = get_skill_list(df_skill['Etc Skill'])
skill_list5 = get_skill_list(df_skill['Knowledge'])
## skill_list는 모든 사람이 가지고 있는 스킬들(중복제거)
skill_list_un1 = set(skill_list1)
skill_list_un2 = set(skill_list2)
skill_list_un3 = set(skill_list3)
skill_list_un4 = set(skill_list4)
skill_list_un5 = set(skill_list5)
```

```python
## skill별 Top 10
print(pd.Series(skill_list1).value_counts().iloc[:10])
print(pd.Series(skill_list2).value_counts().iloc[:10])
print(pd.Series(skill_list3).value_counts().iloc[:10])
print(pd.Series(skill_list4).value_counts().iloc[:10])
print(pd.Series(skill_list5).value_counts().iloc[:10])
```

```python
## 가지고 있는 스킬의 수를 구하는 함수
def get_cnt_skill(ro):
    return len(ro) 
```

```python
df_skill['endorsement cnt'] = df_skill['Skill'].apply(lambda x:get_cnt_skill(x))
df_skill['tool cnt']        = df_skill['Tool'].apply(lambda x:get_cnt_skill(x))
df_skill['personal cnt']    = df_skill['Personal Skill'].apply(lambda x:get_cnt_skill(x))
df_skill['etc cnt']         = df_skill['Etc Skill'].apply(lambda x:get_cnt_skill(x))
df_skill['knowledge cnt']   = df_skill['Knowledge'].apply(lambda x:get_cnt_skill(x))
```

```python
def have_skill(ski,ser):
    tf=0
    for idx in range(len(ser)):
        if ski in ser.iloc[idx]:
            tf+=1
    return tf
```

```python
## 스킬 컬럼을 선택했을때 갖고 있는 스킬의 분포를 확인하고 싶다.
def get_dist(col,tp_sel):
    plt.figure(figsize = (10,5))
    a=list()
    for idx in range(len(df_skill[col])):
        for i,v in enumerate(df_skill[col].iloc[idx]):
            a.append(v)
    return pd.Series(a).value_counts().iloc[:tp_sel].plot(kind='bar')
        
        
    #return df_skill[col].value_counts()
```

## Years Experienced

```python
df_exp = df[['Experience']]
```

```python
def get_year(tr):
    mo_tr = 0
    ye_tr = 0
    year = 0
    for i in range(len(tr)):
        
        tr[i]['Years'] = tr[i]['Years'].replace(' ','').replace('yrs','yr').replace('mos','mo')
        if 'yr' in tr[i]['Years'] and 'mo' in tr[i]['Years']:
            ye_tr+=np.int(tr[i]['Years'].replace('mo','').split('yr')[0])
            mo_tr+=np.int(tr[i]['Years'].replace('mo','').split('yr')[1])
        elif 'yr' not in tr[i]['Years'] and 'mo' in tr[i]['Years']:
            mo_tr+=np.int(tr[i]['Years'].replace('mo',''))
            ye_tr+=0
        elif 'yr' in tr[i]['Years'] and 'mo' not in tr[i]['Years']:
            ye_tr+=np.int(tr[i]['Years'].replace('yr',''))
            mo_tr+=0
        else:
            ye_tr+=0
            mo_tr+= 0
    year = round(((ye_tr*12)+mo_tr)/12,1)
    return year
```

```python
df_exp['Years'] = df_exp['Experience'].apply(lambda x:get_year(x))
```
