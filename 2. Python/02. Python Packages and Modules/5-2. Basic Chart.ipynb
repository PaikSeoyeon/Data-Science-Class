# matplotlib을 이용한 데이터 시각화 프로그래밍

```python
import matplotlib.pyplot as plt

X = range(100)
Y = [value **2 for value in X]

plt.plot(X,Y)
plt.show()
```

### 넘파이 사용

```python
import math
import matplotlib.pyplot as plt

T = range(100)
X = [(2 * math.pi * t) / len(T) for t in T]
Y = [math.sin(value) for value in X]

plt.plot(X,Y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(0, 2*np.pi, 100)
Y = np.sin(X)

plt.plot(X,Y)
plt.show()
```

### 넘파이 패키지

```python
X = np.linspace(-3, 2, 200)
Y = X**2 - 2*X + 1.

plt.plot(X,Y)
plt.show()
```

### 다중 곡선 그리기

```python
X  = np.linspace(0, 2*np.pi, 100)
Ya = np.sin(X)
Yb = np.cos(X)

plt.plot(X, Ya)
plt.plot(X, Yb)
plt.show()
```

```python
X  = np.linspace(0, 2*np.pi, 10)
Ya = np.sin(X)
Yb = np.cos(X)

plt.plot(X, Ya)
plt.plot(X, Yb)
plt.show()
```

```python
def plot_slope(X, Y):
    Xs = X[1:] - X[:-1]
    Ys = Y[1:] - Y[:-1]
    plt.plot(X[1:], Ys/Xs)
    
X = np.linspace(-3, 3, 100)
Y = np.exp(-X**2)

plt.plot(X, Y)
plot_slope(X, Y)
plt.show()
```

### 파일 데이터로부터 곡선 그리기

```python
f = open('my_data.txt', 'w')
f.write('0  0')
f.write('\n1  1')
f.write('\n2  4')
f.write('\n4 16')
f.write('\n5 25')
f.write('\n6 36')
f.close()

f = open('my_data.txt')
txt= f.read()
print(txt)
f.close()
```

```python
X, Y = [], []
for line in open('my_data.txt', 'r'):
    values = [float(s) for s in line.split()]
    X.append(values[0])
    Y.append(values[1])
    
plt.plot(X,Y)
plt.show
```

```python
with open('my_data.txt', 'r') as f:
    X,Y = zip(*[[float(s) for s in line.split()] for line in f])
    
plt.plot(X,Y)
plt.show()
```

### 다중 막대 차트 그리기

```python
data = [[5., 25., 50., 20.],
        [4., 23., 51., 17.],
        [6., 22., 52., 19.]]

X = np.arange(4)
plt.bar(X + 0.00, data[0], color='b', width=0.2)
plt.bar(X + 0.2, data[1], color='g', width=0.2)
plt.bar(X + 0.4, data[2], color='r', width=0.2)

plt.show()
```

```python
len(data)
```

```python
data = [[5., 25., 50., 20.],
        [4., 23., 51., 17.],
        [6., 22., 52., 19.]]

color_list = ['b', 'g', 'r']
gap = .8 / len(data)
for i, row in enumerate(data):
    X = np.arange(len(row))
    plt.bar(X + i*gap, row, 
           width = gap, 
           color = color_list[i % len(color_list)])

plt.show()
```

```python
# 값을 두 집합 이상으로 분할

A = np.array([5., 30., 45., 22.])
B = np.array([5., 25., 50., 20.])
C = np.array([1.,  2.,  1.,  1.])
X = range(4)

plt.bar(X, A, color='b')
plt.bar(X, B, color='g', bottom=A)
plt.bar(X, C, color='r', bottom=A+B)
plt.show()
```

```python
data = np.array([[5., 30., 45., 22.],
                 [5., 25., 50., 20.],
                 [1.,  2.,  1.,  1.]        
    ])

color_list = ['b', 'g', 'r']

X = np.arange(data.shape[1])

for i in range(data.shape[0]):
    plt.bar(X, data[i],
           bottom = np.sum(data[:i], axis=0),
           color = color_list[i % len(color_list)])

plt.show()
```

### 삼각화 그리기

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.tri as tri

data = np.random.rand(100, 2)

triangles = tri.Triangulation(data[:,0], data[:,1])

plt.triplot(triangles)
plt.show()
```
