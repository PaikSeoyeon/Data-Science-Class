```
### 활성화 함수
### Sigmoid, ReLu, leaky ReLu, tanh, 
```

```
import numpy as np
import matplotlib.pyplot as plt
```

```
x = np.linspace(-5, 5, 300) # -5~5 범위에서 균일 간격으로 300개.

## step function : 0보다 작으면 0, 크면 1
step_fnc = x.copy()
step_fnc[step_fnc <= 0] = 0
step_fnc[step_fnc > 0] = 1
print(step_fnc)

## sigmoid function : 초기에 사용된 활성화 함수. (0~1)
sig_fnc = 1/(1+np.exp(-x))
print(sig_fnc)

## relu function : 0보다 작으면 0, 0보다 크면 비례하게 늘어남.
relu_fnc = np.maximum(x, 0)
print(relu_fnc)

## tanh_fnc : 값의 범위 (-1~1)
tanh_fnc = np.tanh(x)
print(tanh_fnc)
```

```
plt.rcParams["figure.figsize"] = (8,6)

plt.subplot(2,2,1)
plt.plot(step_fnc)

plt.subplot(2,2,2)
plt.plot(sig_fnc)

plt.subplot(2,2,3)
plt.plot(relu_fnc)

plt.subplot(2,2,4)
plt.plot(tanh_fnc)
```

```
# 음수 표시
import matplotlib
matplotlib.rcParams['axes.unicode_minus'] = False
```

```
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
ax1.plot(x, step_fnc)
ax2.plot(x, sig_fnc)
ax3.plot(x, relu_fnc)
ax4.plot(x, tanh_fnc)
```

```
plt.plot(x, tanh_fnc, label='tanh')
plt.plot(x, step_fnc, label='relu')
plt.plot(x, sig_fnc, label='sigmoid')
plt.plot(x, relu_fnc, label='step function')
plt.legend(loc='best')
plt.xlabel('x')
plt.ylabel('relu(x), tanh(x)')
```
