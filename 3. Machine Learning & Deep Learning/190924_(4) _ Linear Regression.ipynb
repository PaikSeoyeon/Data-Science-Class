```
import numpy as np
import sys
import pandas as pd
import scipy as sp
import seaborn as sns 
import matplotlib.pyplot as plt
import mglearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

### Boston 데이터 셋을 활용한 회귀 모델 만들어보기


```
from sklearn.datasets import load_boston
```

```
boston = load_boston()

print(boston.keys())
print(boston.items())
```

### 데이터 셋 나누기

```
X = boston.data
Y = boston.target
X_train, X_test, y_train, y_test = train_test_split(X, Y, train_size = 0.8,
                                                                                                        random_state=77)
```

### 모델 학습

```
model = LinearRegression().fit(X_train, y_train)
```

### 모델 예측

```
pred = model.predict(X_test)
pred
```

```
dat = pd.DataFrame({"y_test" : y_test,
                    "prediction" : pred,
                    "diff" : y_test-pred})
dat
```

### MAE 구하기 - MSE / MAE / RMSE / RMLSE

```
sum(abs(dat['diff']))/len(dat['diff'])
```

```
dat['diff'] ** 2
```

### MSE 구하기
(실제값-예측값)^2 의 합을 데이터의 샘플 수로 나누어 준 것

```
mse_value = sum(dat['diff']**2) / len(dat['diff'])
mse_value
```

### RMSE 구하기
MSE 에 루트

```
result = np.sqrt(mse_value)
result
```

### R^2 값 구하기 (결정계수)

```
print('훈련 데이터 세트 점수 : {:.2f}' .format(model.score(X_train, y_train)))
print('테스트 데이터 세트 점수 : {:.2f}' .format(model.score(X_test, y_test)))
```
