```
# pandas 라이브러리 import
import pandas as pd
```

```
# 데이터 불러오기
df = pd.read_csv('bike_train.csv')

# 데이터 확인
df.info()
```

```
df.head()
```

```
# 데이터 컬럼 나누기
# data에는 예측하고자 하는 'count' -> 제외
# 'casual', 'registered' 과적합 발생 -> 제외
# datetime 의 type : object -> 제외

data = df.drop(['count', 'datetime', 'casual', 'registered'], axis=1)
```

```
target = df['count']
```

```
# sklearn 라이브러리에서 train_test_split 불러오기
from sklearn.model_selection import train_test_split
```

```
# train_test_split 으로 train 데이터와 test 데이터로 나누기
X_train, X_test, y_train, y_test = train_test_split(data, target,
                                                    random_state=0)
```

```
# sklearn에서 LinearRegression 불러오기
from sklearn.linear_model import LinearRegression
```

```
# train 데이터로 LinearRegression 모델 만들기
model = LinearRegression().fit(X_train, y_train)
```

```
# 만들어진 모델로 test 데이터 예측하기
pred = model.predict(X_test)
pred
```

```
# 보기 좋게 데이터 프레임으로 만들기
dat = pd.DataFrame({'y_test' : y_test, 'prediction' : pred, 'diff' :y_test-pred})
dat
```

```
# MAE(mean absolute error, 평균 절대값 오차)
sum(abs(dat['diff']))/len(dat['diff'])
```

```
# MSE(meand square error, 평균 제곱 오차)
mse_value = sum(dat['diff']**2)/len(dat['diff'])
mse_value
```

```
# RMSE(root mean square error, 평균 제곱근 오차)
import numpy as np
result = np.sqrt(mse_value)
result
```

```
# 훈련 데이터 셋과 테스트 데이터 셋 점수 확인
print('훈련 데이터 셋 점수 : {:.2f}' .format(model.score(X_train, y_train)))
print('테스트 데이터 셋 점수 : {:.2f}' .format(model.score(X_test, y_test)))
```
