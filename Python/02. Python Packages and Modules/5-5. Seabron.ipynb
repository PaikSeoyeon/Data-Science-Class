### boxplot

```python
plt.rcParams['figure.figsize'] = (12,6)

sns.set_style("whitegrid")
sns.set()
```

```python
tips = sns.load_dataset("tips")
tips.head(5)
```

```python
sns.boxplot(x=tips["total_bill"])
```

```python
tips.day.unique()
```

```python
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips, palette="Set3")
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", hue="time", data=tips, linewidth=0.9, palette="Set1")
plt.show()
```

```python
sns.boxplot(x="time", y="tip", data=tips, order=["Dinner", "Lunch"])
plt.show()
```

### kdeplot

```python
import seaborn as snsn; sns.set(color_codes=True)
```

```python
import numpy as np; np.random.seed(10)
```

```python
mean, cov = [0,2], [(1,.5),(5,1)]
```

```python
mean
```

```python
cov
```

```python
x, y = np.random.multivariate_normal(mean, cov, size=50).T
```

```python
x
```

```python
y
```

```python
plt.plot(x)
```

```python
sns.kdeplot(x)
plt.show()
```

```python
sns.kdeplot(x, shade=True, color="r")
plt.show()
```

```python
sns.kdeplot(x, y)
plt.show()
```

```python
sns.kdeplot(x, y, shade=True)
plt.show()
```

```python
sns.kdeplot(x, y, n_levels=50, cmap="Purples_d")
plt.show()
```

```python
sns.kdeplot(x, bw=.1)
plt.show()
```

```python
sns.kdeplot(y, vertical=True)
plt.show()
```

```python
sns.kdeplot(x, bw=.15, cut=10)
plt.show()
```
