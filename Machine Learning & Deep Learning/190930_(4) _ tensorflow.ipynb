## 텐서 플로워의 자료형
* 플레이스 홀더 : 나중에 값을 입력 받기 위해 사용되는 매개변수 (parameter)
* shape(?,3) 두번째 차원 요소가 3이다. 앞의 차원은 나중에 지정

```
tensorX = tf.placeholder(tf.float32, [None, 2])
```

```
# X 에 넣을 데이터 초기화
x_data = [[1,1], [2,2]]
```

```
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```
sess.run(tensorX, feed_dict={tensorX:x_data})
sess.close()
```

## 행렬 연산

```
tensorX = tf.placeholder(tf.float32, [None,2])

W = tf.Variable(tf.random_normal([2,2]))
b = tf.Variable(tf.random_normal([2,1]))

print(W)
print(b)
```

```
y = tf.matmul(tensorX, W) + b
y
```

```
### x_data, W, b, y 값 출력해보기
print("x_data : \n{}\n" .format(x_data))

sess = tf.Session()
sess.run(tf.global_variables_initializer())
print("W : \n{}\n" .format(sess.run(W)))
print("b : \n{}\n" .format(sess.run(b)))
print("y : \n{}\n" .format(sess.run(y, feed_dict={tensorX:x_data})))

sess.close()
```

# MNIST

```
import tensorflow as tf
```

```
mnist = tf.keras.datasets.mnist
print(mnist)
```

```
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```
y_train[0:3]
```

```
x_train[0:3]
```

```
# 값의 정규화
# Dropout, 값의 정규화(0~1)
x_train, x_test = x_train/255.0, x_test/255.0
```

## tf 신경망 만들기

```
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28,28)),
  tf.keras.layers.Dense(512, activation=tf.nn.relu),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])
```

```
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```
model.fit(x_train, y_train, epochs=5)
```

```
model.evaluate(x_test, y_test)
```
